# Agrupamento com C-Means.

```python
# Importando bibliotecas:

import numpy as np
import skfuzzy
from sklearn import datasets
from sklearn.metrics import confusion_matrix

# pip install - U scikit-fuzzy (executar no Anaconda prompt)
```

```python
# Carregando da base de dados iris, que ja esta disponivel no sklearn:

iris = datasets.load_iris()
```

```python
# Aplicação do algoritimo definindo três cluster (c = 3) e passando a matriz transposa (iris.data.T).

# Os outros parâmetros são obrigatórios e são os default indicados na documentação!

r = skfuzzy.cmeans(data = iris.data.T, c = 3, m = 2, error = 0.005, maxiter = 1000, init = None)
```

```python
# Obtendo as porcentagens de um registro pertencer a um cluster, que está na posição 1 da matriz retornada:

previsoes_porcentagem = r[1]
```

```python
# Visualização da probabilidade de um registro pertencer a cada um dos cluster (a somatória é 1.0 que indica 100%)

for x in range(150):
    print( previsoes_porcentagem[0][x] , previsoes_porcentagem[1][x], previsoes_porcentagem[2][x] )

```

```python
# Geração da matriz de contingência para comparação com as classes originais da base de dados:

previsoes = previsoes_porcentagem.argmax(axis = 0)
resultados = confusion_matrix(iris.target, previsoes)
resultados
```
